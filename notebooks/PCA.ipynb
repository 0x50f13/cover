# PCA
Pricnipal component analysis allows to reduce data dimension with losing the least amount of information. Here we will use neurons selected by either mutual or spatial information methods from `best_neurons` notebook.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
atRoot = 'atRoot' in vars()
if not atRoot:
    os.chdir("..")
atRoot = True
```

```python
DATASET="data/22ht1_normalized.csv" # Dataset path
SCORED_SI=[541, 522, 309, 489, 203, 488, 521, 427, 54, 544]
SCORED_MI=[553, 551, 527, 549, 142, 470, 489, 401, 155, 547]
data=pd.read_csv(DATASET)
```

```python
def do_pca(scored):
    X=[list(data.iloc[:,x]) for x in scored]
    X=np.array(X)
    X=X.T
    pca=PCA(n_components=2)
    return pca.fit_transform(X)
```

```python
transformed_si=do_pca(SCORED_SI)
```

```python
transformed_mi=do_pca(SCORED_MI)
```

```python
def plot_transformed(transformed):
    vX=[x for [x,_]  in transformed]
    vY=[y for [_,y]  in transformed]
    plt.scatter(vX,vY)
    plt.show()
```

```python
plot_transformed(transformed_si)
plot_transformed(transformed_mi)
```

```python
F_SCORED_MI=[527, 320, 384, 90, 23, 418, 103, 403, 201, 447]
plot_transformed(do_pca(F_SCORED_MI))
```
